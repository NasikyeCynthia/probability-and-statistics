```python
from scipy.stats import chi2_contingency  
```

```python
CrosstabResult=pd.crosstab(index=data['State'],columns=data['Sub_Category'])
print(CrosstabResult)
test = chi2_contingency(CrosstabResult)
test
```

p_value is 0.0 therefore state ans sub_category are correlated

```python
#statistical test between one continuous and one categorical variable   
from scipy.stats import f_oneway
def ccat(data, Unit_Cost, Sub_Category):
    Group_data = [
            group[Unit_Cost].values for _, group in data.groupby(Sub_Category)
    ]
    f_starts,p_value=f_oneway(*Group_data)
    if p_value > 0.05:
        print(f"{Unit_Cost} is not correlated with {Sub_Category} with a p_value of {p_value}")
    elif p_value < 0.05:
        print(f"{Unit_Cost} is correlated with {Sub_Category} with a p_value of {p_value}")
        
ccat(data, 'Unit_Cost','Sub_Category')

        
```
